```python
def customer_service_bot():
    """
    A simple rule-based chatbot for a customer service scenario.
    """
    rules = {
        "hello": "Hello! Welcome to customer service. How can I assist you today?",
        "hi": "Hello! Welcome to customer service. How can I assist you today?",
        "shipping": "You can track your order status at www.example.com/tracking.",
        "order status": "You can track your order status at www.example.com/tracking.",
        "track": "You can track your order status at www.example.com/tracking.",
        "return": "Our policy allows returns within 30 days of purchase. Visit www.example.com/returns for more info.",
        "refund": "Our policy allows returns within 30 days of purchase. Visit www.example.com/returns for more info.",
        "hours": "We are open from 9 AM to 6 PM, Monday to Friday.",
        "open": "We are open from 9 AM to 6 PM, Monday to Friday.",
        "agent": "If you'd like to speak with a human agent, please call us at 1-800-555-0199.",
        "human": "If you'd like to speak with a human agent, please call us at 1-800-555-0199.",
        "bye": "Thank you for visiting. Have a great day!",
        "exit": "Thank you for visiting. Have a great day!"
    }

    print("Chatbot: Hello! Welcome to customer service. Type 'bye' to exit.")

    while True:
        user_input = input("You: ").lower()

        if user_input in ["bye", "exit"]:
            print(f"Chatbot: {rules[user_input]}")
            break

        response_given = False
        for keyword, response in rules.items():
            if keyword in user_input:
                print(f"Chatbot: {response}")
                response_given = True
                break

        if not response_given:
            print("Chatbot: I'm sorry, I don't understand. Please ask about 'shipping', 'return', or 'hours'.")

# Run the chatbot
customer_service_bot()
```
